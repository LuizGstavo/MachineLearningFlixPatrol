```python
import pandas as pd
from pycaret.classification import *
# Carregando os dados
data = pd.read_csv('creditcard.csv')
# Configurando o ambiente
setup_clf = setup(data, target='Class',
normalize=True,
fix_imbalance=True, # Corrige desbalanceamento de classes
session_id=42)
# Comparando modelos
compare_models(sort='Recall') # Priorizando recall
```

```python
best_models = compare_models(n_select=3, sort='Recall')
# Selecionando o melhor modelo
best_model = create_model(best_models)
# Ajustando hiperparâmetros
tuned_model = tune_model(best_model, optimize='Recall')
# Finalizando o modelo (blend dos melhores)
final_model = ensemble_model(tuned_model)
# Interpretando o modelo
interpret_model(final_model)
```

```python
# Salvando o modelo
save_model(final_model)
```
